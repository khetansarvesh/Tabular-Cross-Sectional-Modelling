# CONVERTING DICTIONARY INTO DATAFRAME
Data in form of table can be thought of dictionary at the backend.

```
dic = {
    "Name" : ["Sarvesh Khetan","Gorang Agarwal","Kazi","Vineeth Varma"],
    "Marks" : [8.3,7.7,8,7.7],
    "Age" : [29,52,21,65],
    "City" : ["Mumbai","Delhi","Kolkata","Hyderabad"]
}
```

```
l = dic.keys()  #this is a dictionary function and gives us all the list of keys(column_names) present in the dataset
print(l)
```

```
l = dic.values()
print(l)
```

```
df = pd.DataFrame(dic)
df
```

In actual senario you will be provided with this row/column kind of data but at the backed it is stored in dictionary format

```
l = df.keys()
print(l)
#same result hence it proves that both are same thing only,just representation is different
#df.columns   #very imp,dic.keys() was a function but df.columns is not a function call ,it also returns same thing what keys() returned i.e a list of  keys(column_names)
#this small concept will be very helpful later
```

```
l = df.values   #very important to note that with dic we used values() while with df we used values
print(l)
```

```
df.shape #gives (no of row,no of column) as a tuple datatype
#again this is not a function call cause we dont use ()
```

```
df.info()  #gives names of all columns along with its datatype
#object means string datatype
#since () used means it is a function call
```

now we will see how to get any particular data based on some conditions that we pass

```
df['Age'] >= 30  #now we know that df['Age'] gives all the rows that are in column Age
#so df['Age'] >= 30 must give those rows in column age satisfies this condition,and it does but in boolean format
#return type is a dictionary cause you got a matrix back
```

to get the actual data we need to map this boolean as shown below

```
filt = df['Age'] >= 30  #gives true false boolean return in form of generator
df.loc[filt]  #mapping  #you get exactly same result if you use df[filt] but an error if you use df.iloc(filt)
#return type is a dictionary cause you got a matrix back
```

```
df.loc[filt,['Name','Marks']]
#return type is a dictionary cause you got a matrix back
```

now with this filter you can use all kind of logical operators for columns wuth int/float data type

```
filt2 = (df['Age'] < 50) & (df['Marks'] > 6)
df.loc[filt2]
#return type is a dictionary cause you got a matrix back
```

for columns with string data types you have some other set of function available called string function to carry out all the operations successfully but one cannot make functions for everything hence one needs to know how these functions can be made

```
cities = ["Mumbai","Kolkata","Banglore"]
filt = df['City'].isin(cities)      #using this predifined function here but below we will try to make our own such function
print(filt)
df.loc[filt]
#return type is a dictionary cause you got a matrix back
```

```
#implementing samae above logic without using inbuild function
#to implement this you need to know the fact that table is store as a dictionary at backene
cities = ["Mumbai","Kolkata","Banglore"]
filt=[]
for i in df['City']:    #or use dic["City"] you get the same result cause at the backend it is a dictionary
  c=0
  for j in cities:     #cause we know that df["City"] gives series datatype which is a generator
    if i == j:
      c=1
      break

  if c == 1:
    filt.append(True)
  else:
    filt.append(False)

print(filt)
df.loc[filt]
#see you get exactly the same result...very imp,no one teaches this
#hence now no need to remember all the other functions,just code your own logic
#return type is a dictionary cause you got a matrix back
```

Pandas allows us two sort wrt two colums also i.e. of two values are same in column 1 then it sorts according to column 2

```
df.sort_values(by=['Marks',"Age"],ascending=[True,False],inplace=True)
df  #first sort occurs in ascending order wrt marks but if marks are same then sort occurs
# in descending order wrt age
#similarly you can pass as many no of columns as you want in the list
```

## Updating Single Column Data/Multiple rows
First you need to extract that column

```
#let us change all the names to caps
l1 = []
for i in df['name']:    #same thing can also be done using df["name"]=df["name"].str.lower() function
 l1.append(i.upper())
df["name"] = l1 
df
```

```
#another way to do same above thing is to use apply() funtion.This is a higher order fuction and takes a user defined function as an argument 
def update_name(j):
  return j.upper()
df["CITY"]=df["CITY"].apply(update_name)    #You should use lambda functions instead
df
#what apply does is it passes individual element of df["name"] to the funtion,hence you need not write the for loop
#no need to use this,just keep in mind so that if someone else uses it you should understand
```

```
#alternative to apply() is map()...both have exactly same funtionality
def update_name(j):
  return j.lower()
df["CITY"]=df["CITY"].map(update_name)    #You should use lambda functions instead
df
```

# ADDING ROWS AND COLUMNS
This is used to add row(s) or column(s)

### Adding Single New Column


```
l = []                      #first you need to get the data that you want to add
for i in df["name"]:
  str=''
  str = str + i[0] + '.'
  for j in i:
    if(j == ' '):
      a = i.find(j)
      str = str + i[a+1] + '.'
      break
  l.append(str)
print(l)      
```

```
df["Initials"] = l    #now simply make a column and pass the data
df
```
