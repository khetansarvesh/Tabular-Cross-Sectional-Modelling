
### SECTION 1 : Multi Line Plot on Same Axes

always remember that show() command should be the last command always for obvious reasons. Multiple such colours, linestyle and markers are available,refer doc for same

```
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("PLOT",fontsize = 20) #fontsize is optional
plt.plot(x1,y1,label = "first",color="red",linestyle="--",marker='o',alpha = 0.5) #this already plotted to graph at the backend
plt.plot(x2,y2, label = "second",color="#444444",linestyle="-.",linewidth=10)
plt.legend()  #you have added the legend but this command is used to show that legend
plt.grid(True) #to add grids
plt.fill_between(x2,y2,color="green",alpha=0.3) #fill used to fill the space #alpha means opacity
plt.show()  #to show what was plotted at backend
```

As we can conclude from above graph that x2 is exactly linearly dependent on y2 while x1 is slightly linearly dependent on y1. 
Now in above graph fill happens about x axis(i.e y=0) but we can change that and specify about what axis the fill should occur

```
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("PLOT",fontsize = 20) 
plt.plot(x1,y1,label = "first",color="red",linestyle="--",marker='o') 
plt.plot(x2,y2, label = "second",color="#444444",linestyle="-.",linewidth=10)
plt.legend()
plt.grid(True)
plt.fill_between(x2,y2,7000,color="green",alpha=0.3)
plt.show()
```

Now we have many build in method to improve look of our plot which are listed below.Now you can also save this image using plt.savefig("fle_name") command.Write this above show() command

```
print(plt.style.available) #to print all available styles
plt.style.use('fivethirtyeight')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("PLOT",fontsize = 20) #fontsize is optional
plt.plot(x1,y1,label = "first",color="red",linestyle="--",marker='o') #this already plotted to graph at the backend
plt.plot(x2,y2, label = "second",color="#440094",linestyle="-.",linewidth=10)
plt.legend()  #you have added the legend but this command is used to show that legend
plt.grid(True) #to add grids
plt.tight_layout()#used to add padding
plt.show()
```

```
fig , ax = plt.subplots() #by default it returns fig = 1 and ax = 1
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_title("PLOT",fontsize = 20) #fontsize is optional
ax.plot(x1,y1,label = "first",color="red",linestyle="--",marker='o',alpha = 0.5) #this already plotted to graph at the backend
ax.plot(x2,y2, label = "second",color="#444444",linestyle="-.",linewidth=10)
ax.legend()  #you have added the legend but this command is used to show that legend
ax.grid(True) #to add grids
ax.fill_between(x2,y2,color="green",alpha=0.3) #fill used to fill the space #alpha means opacity
plt.show()  #to show what was plotted at backend
#hence you can see we got the same thing as before
```

```
fig , ax = plt.subplots(nrows = 2 , ncols=1) #now we are telling that we need 1row and 2 column axes and since we are not specifing fig it gives 1 by default
print(ax)
ax[0].set_xlabel("x axis") #ax[0] refers to row 1 column 1 ax
ax[0].set_ylabel("y axis")
ax[0].set_title("11",fontsize = 20) #fontsize is optional
ax[1].set_xlabel("x axis") #ax[1] refers to row 2 column 1 ax
ax[1].set_ylabel("y axis")
ax[1].set_title("21",fontsize = 20) #fontsize is optional

ax[0].plot(x1,y1,color="red",linestyle="--",marker='o',alpha = 0.5) #this already plotted to graph at the backend
ax[1].plot(x2,y2,color="#444444",linestyle="-.",linewidth=10)

ax[0].grid(True) #to add grids
ax[1].grid(True)

ax[1].fill_between(x2,y2,color="green",alpha=0.3) #fill used to fill the space #alpha means opacity

plt.tight_layout()
plt.show()  #to show what was plotted at backend

#similarly if you had 2 rows and 2 columns then 
# [0,0] to refer top left ax
# [0,1] to refer top right ax
# [1,0] to refer bottom left ax
# [1,1] to refer bottom right ax
```

```
fig1 ,ax1 = plt.subplots() #now we are telling that we need two figures
fig2 , ax2 = plt.subplots()

ax1.set_xlabel("x axis")
ax1.set_ylabel("y axis")
ax1.set_title("PLOT",fontsize = 20) #fontsize is optional
ax2.set_xlabel("x axis")
ax2.set_ylabel("y axis")
ax2.set_title("PLOT",fontsize = 20) #fontsize is optional


ax1.plot(x1,y1,color="red",linestyle="--",marker='o',alpha = 0.5) #this already plotted to graph at the backend
ax2.plot(x2,y2,color="#444444",linestyle="-.",linewidth=10)

ax1.grid(True) #to add grids
ax2.grid(True)

ax2.fill_between(x2,y2,color="green",alpha=0.3) #fill used to fill the space #alpha means opacity

plt.tight_layout()
plt.show()  #to show what was plotted at backend
```

# **HIGHER DIMENSION PLOTTING (4D / 5D ...) :**

Only line or scatter plot in 4D/5D... makes sense


Now we know that it is not possible to plot 4d or above. Hence we cannot plot those, but we can kind of represent that 4d/5d/... plots as 2d plots using a concept called **principal component analysis (pca)** which can be implimented using sklearn library.

PCA is used to convert N columns into K columns and then plot that k columns but here since we can plot only 2D plots our k will always be 2 irrespective of how much variance that 2 principal components accounts for.

## LINE PLOT :

```
import pandas as pd
dic_line = {
        'x1' : [1,2,3,4,5,6],
        'x2' : [1,2,3,4,5,6],
        'y'  : [1,2,3,4,5,6]
}
df_line = pd.DataFrame(dic_line)  #here we need to convert it into data frame because the sklearn library takes input in dataframe
df_line   #represent 3d line plot
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components = 1)           #saying that we need to convert it into 2d figure
pca.fit(df_line[['x1','x2']])                       #taking the projection
df_line_new = pca.transform(df_line[['x1','x2']])          #asking the projected values and storing in df_line_new
print(df_line_new)                            #returns array datatype
plt.plot(df_line_new,df_line.iloc[:,2])                               
```

## SCATTER PLOT :

```
import pandas as pd
dic_scatter = {
        'x1' : [1,2,3,4,5,6],
        'x2' : [1,2,3,4,5,6],
        'x3'  : [1,2,3,4,5,6],
        'y' : [0,1,0,1,1,1]
}
df_scatter = pd.DataFrame(dic_scatter)  #here we need to convert it into data frame because the sklearn library takes input in dataframe
df_scatter   #represent 3d scatter plot
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)           #saying that we need to convert it into 2d figure
pca.fit(df_scatter)                       #taking the projection
df_line_new = pca.transform(df_scatter)          #asking the projected values and storing in df_line_new
print(df_line_new)                            #returns array datatype
plt.plot(df_line_new[:,0],df_line_new[:,1])  
```
