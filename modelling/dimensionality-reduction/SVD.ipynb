# **Dataset**

```python
df = pd.read_csv('Tabular-Cross-Sectional-Modelling/dataset/Wine.csv')
df
df.shape
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(np.array(X))
X = pd.DataFrame(X)
X
```

# **SVD - Library Implementation**

```python
from sklearn.decomposition import TruncatedSVD
```

```python
# Initialize SVD and fit on the training data
svd = TruncatedSVD(n_components=12)  # Use one less component than the feature count
X_train_svd = svd.fit_transform(X)

# Calculate the cumulative explained variance
cumulative_explained_variance = np.cumsum(svd.explained_variance_ratio_)
print(cumulative_explained_variance)
```

```python
# Hence we can see at principal component 7 we achieve 90% information and hence we will select 7 components
X_train_svd_selected = svd.transform(X)[:, :7]
```

```python
pd.DataFrame(X_train_svd_selected)
```

# **SVD - Custom Implementation**

```python
from scipy.linalg import svd
from numpy import zeros
from numpy import diag
```

```python
# Singular-value decomposition
U, s, VT = svd(X)

# converting sigma from a vector into a diagonal matrix
Sigma = zeros((178, 13))
Sigma[:13, :13] = diag(s)
```

```python
# reduced
Sigma = Sigma[:, :5]
VT = VT[:5, :]
```

```python
# mehod 1 to get reduced dimensinos
T = U.dot(Sigma)
pd.DataFrame(T)
```

```python
# method 2 to get reduced dimensions
T = X.dot(VT.T)
pd.DataFrame(T)
```
