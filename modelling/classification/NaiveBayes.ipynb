# Data Preprocessing

```python
!wget https://github.com/khetansarvesh/Tabular-Cross-Sectional-Modelling/blob/main/dataset/indians-diabetes.csv
```

```python
# Data Reading
df = pd.read_csv('indians-diabetes.csv')
df
```

```python
#performing feature scaling
X = df.iloc[:,:-1]
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X.values))
Y = df.iloc[:,-1]
```

```python
#train test split - not doing any validation split cause naive bayes algorithm does not has any hyperparameter to be tuned
x_train = X.iloc[:600,:]
x_test = X.iloc[600:,:]
y_train = Y[:600]
y_test = Y[600:]
```

# Data Learing using sklearn Library

```python
from sklearn.naive_bayes import GaussianNB
#from sklearn.naive_bayes import MultinomialNB
NB_classifier = GaussianNB()
NB_classifier.fit(np.array(x_train), np.array(y_train));
```

```python
y_test_pred = NB_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_test_pred))
```

```python
from sklearn.model_selection import cross_validate
NB_result = (cross_validate(NB_classifier,X,Y,cv=7,return_train_score=True))
NB_training_accuracy = NB_result['train_score']
NB_testing_accuracy = NB_result['test_score']
print("training accuracy over 7 folds : {}".format(list(NB_training_accuracy)))
print("average training accuracy : {}".format(np.mean(NB_training_accuracy)))
print("")
print("testing accuracy over 7 folds :{}".format(list(NB_testing_accuracy)))
print("average testing accuracy : {}".format(np.mean(NB_testing_accuracy)))
```
